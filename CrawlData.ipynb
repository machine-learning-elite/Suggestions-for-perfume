```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.amazon.com/s?k=perfume&crid=2F1V21B1DWQBK&sprefix=perfume%2Caps%2C380&ref=nb_sb_noss_1'
response = requests.get(url)
```

```python
response
```

```python
if response.status_code == 503:
    soup = BeautifulSoup(response.content, 'html.parser')
    products = soup.find('div', class_="sg-col-inner")
```
